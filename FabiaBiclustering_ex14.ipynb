```python
def plt_fab(X, biclust, cycle):
    m = FabiaBiclustering(n_clusters=biclust, n_iter=cycle)  
    m.fit(X.T)
    # Outer product of 1st column of Z and 1st column of L:
    X0 = np.outer(m.Z_[:, 0], m.L_[0, :]) 
    plt.pcolor(X0, cmap=plt.cm.RdBu)
    plt.show()

plt_fab(X, 4, 50)
```

```python
plt_fab(X, 4, 500)
```

```python
plt_fab(X, 15, 50)
```

```python
plt_fab(X, 15, 500)
```

```python
plt_fab(X, 20, 50)
```

```python
plt_fab(X, 20, 500)
```

### Interpretion of the results:
The visualization of the results show that the high numbers of biclustering/cycles of the rows and columns yield a more compressed culsters than the ones with smaller numbers of biclustering/cycles. 

So the 1st clustering result is the least compressed and the last is the most compressed.
