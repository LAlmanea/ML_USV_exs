```python
import sys
sys.path.append('/usr/local/lib/python3.6/site-packages')

from matplotlib import cm 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline  

from numpy import genfromtxt
data = genfromtxt('faces94.csv', delimiter=',')
```

```python
def plotPCA(img,comp_perc = None):
    # performing pca
    if comp_perc == None:
        x = img.reshape(45,50).T
        plt.imshow(x, cmap="gray")
        #axis("off")
        tight_layout()
    else:
        x = img.reshape(45,50).T
        pca = PCA(n_components= comp_perc)
        pca.fit(x)
        transformed_x = pca.fit_transform(x)
        restored_x = pca.inverse_transform(transformed_x)
        plt.imshow(restored_x, cmap='gray')
        #axis("off")
        tight_layout()
    
plotPCA(data[15])
```

```python
from pylab import *
subplot(2, 3, 1)
plotPCA(data[15],0.5)
title('50% Downprojection')

subplot(2, 3, 2)
plotPCA(data[15],0.75)
title('75% Downprojection')

subplot(2, 3, 3)
plotPCA(data[15],0.9)
title('90% Downprojection')

subplot(2, 3, 4)
plotPCA(data[15],0.95)
title('95% Downprojection')

subplot(2, 3, 5)
plotPCA(data[15],0.99)
title('99% Downprojection')

subplot(2, 3, 6)
plotPCA(data[15])
title('Original Image')
```
