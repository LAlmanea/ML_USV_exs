```python
import sys
sys.path.append('/usr/local/lib/python3.6/site-packages')

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FastICA
%matplotlib inline

data = np.load("cifar10.npy") 
```

```python
def plotImg(x):
    fig, ax = plt.subplots(figsize=(4, 4))
    x1 = x.reshape((3, 32, 32))
    x2 = np.rollaxis(x1, 0, 3)
    x3 = (x2 - x2.min()) / x2.ptp()
    img = ax.imshow(x3, interpolation="none")
    ax.grid(False)
    ax.axis("off")
    fig.subplots_adjust(left=0, top=1, bottom = 0, right=1)

plotImg(data[0])
```

```python
# Apply PCA
pca = PCA(n_components=100)
pca.fit(data)

# Plot the first 10 principal components
plotImg(pca.components_[0])
plotImg(pca.components_[1])
plotImg(pca.components_[2])
plotImg(pca.components_[3])
plotImg(pca.components_[4])
plotImg(pca.components_[5])
plotImg(pca.components_[6])
plotImg(pca.components_[7])
plotImg(pca.components_[8])
plotImg(pca.components_[9])


# Plot the last 10 principal components
plotImg(pca.components_[-1:])
plotImg(pca.components_[-2:-1])
plotImg(pca.components_[-3:-2])
plotImg(pca.components_[-4:-3])
plotImg(pca.components_[-5:-4])
plotImg(pca.components_[-6:-5])
plotImg(pca.components_[-7:-6])
plotImg(pca.components_[-8:-7])
plotImg(pca.components_[-9:-8])
plotImg(pca.components_[-10:-9])
```

```python
# Apply ICA
ica = FastICA(n_components=100)
ica.fit(data)

#Plot the first 10 independent components
plotImg(ica.components_[0])
plotImg(ica.components_[1])
plotImg(ica.components_[2])
plotImg(ica.components_[3])
plotImg(ica.components_[4])
plotImg(ica.components_[5])
plotImg(ica.components_[6])
plotImg(ica.components_[7])
plotImg(ica.components_[8])
plotImg(ica.components_[9])

#Plot the last 10 independent components
plotImg(ica.components_[-1:])
plotImg(ica.components_[-2:-1])
plotImg(ica.components_[-3:-2])
plotImg(ica.components_[-4:-3])
plotImg(ica.components_[-5:-4])
plotImg(ica.components_[-6:-5])
plotImg(ica.components_[-7:-6])
plotImg(ica.components_[-8:-7])
plotImg(ica.components_[-9:-8])
plotImg(ica.components_[-10:-9])
```

. Describe what you see: how does PCA differ from ICA?

PCA: The first 10 principal components show the maximum variance of brightness/darkness in the images. Whereas the 10 last principal components show mostly noise, as the PCA algorithm tries to fit into fine details of the image.

ICA: The first 10 irincipal components retain similar information to the the last 10 principal components, that seem to show edges and background colours which don't corrolate; hence they're extracted from indipendent components. 

. With PCA, you will notice that the first few components extract a different type of information than the last few. Why do you think that is?

PCA extracts different information for different principal components, because it maximizes for variance in order. So the the first few principal components show the most variance in the data and retain most information, but the last principal components show the lowest variance and retain the least information.

. With ICA, this is not the case. Why?

Because the ICA extracts statistically independent components, and they're not necessarily ordered, so the last few independent components might extract similar information to the first independent components.
